```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('../data/merged_centril_alcohol.csv')

alcohol = df['alcohol_consumption']
centril = df['Cantril ladder score']

plt.figure(figsize=(6,6))
plt.scatter(alcohol, centril,           # x- en y-waarden
            s=50,          # grootte van de punten
            alpha=0.7,     # transparantie
            edgecolor='k') # randkleur

plt.xlabel('Total pure alcohol consumption per capita in liters')
plt.ylabel('Centril ladder score')
plt.title('Alcohol vs Happiness')

plt.grid(True, linestyle='--', alpha=0.3)

m, b = np.polyfit(alcohol, centril, deg=1)  # slope en intercept
plt.plot(alcohol, m*alcohol + b, color='green', lw=2, label=f'y={m:.2f}x+{b:.2f}')
plt.legend()
plt.show()
```

```python
from scipy.stats import pearsonr
#correlation
r, p_value = pearsonr(alcohol, centril)
print(f"Pearson r = {r:.3f}, p-value = {p_value:.3f}")
```
